```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("D:\\work\\Python\\python_for_edureka_doc\\Project\\uv2towea1d\\OnlineNewsPopularity.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = data.iloc[:,1:60]
X.head()
```

```python
from matplotlib.pylab import rcParams
corr = X.corr()
sns.heatmap(corr)
rcParams['figure.figsize'] = 20, 20
```

```python
Y = data["shares"]
Y.head()
```

```python
train_x, test_x, train_y, test_y = train_test_split(X, Y, random_state = 10, test_size = 0.30)
```

```python
lin_model = LinearRegression()
lin_model.fit(train_x, train_y)
```

```python
predicted = lin_model.predict(test_x)
predicted
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(predicted, test_y)
```

```python
plt.plot(predicted, test_y)
plt.show()
```
